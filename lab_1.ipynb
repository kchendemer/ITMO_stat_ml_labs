## 1.1 Вероятностное пространство, формула Байеса

#### 1. Определить(с обоснованием), зависимы или независимы следующие события: <br>
(a) Несовместные события;<br>
(b) События, образующие $\sigma$-алгебру $\Sigma$ в пространств ($\Omega$,$\Sigma$, $P$);<br>
(c) События, имеющие одинаковую вероятность?<br>

#### Ответ:<br>
**(a) Несовместные события** <br>

События A и B являются несовместными, если события не могут произойти одновремено, значит, что $A \cap B = \varnothing \implies P(A \cap B) = 0 $. При $P(A) > 0$ и $P(B) > 0$, имеем 
$$P(A|B) = \frac{P(A \cap B)}{P(B)}$$
Так как А и B несовместные, то они являются взаимо исключающимися, то есть $$P(A|B) = 0$$
При независимости событий вероятность события A не зависит от того, произошло событие B или нет, и в таком случае $$P(A | B) = P(A)$$
Тогда получаем, что несовместные события A и B являются зависимыми,так как $$P(A|B) = 0 \neq P(A)$$

**(b) События, образующие $\sigma$-алгебру $\Sigma$ в пространств ($\Omega$,$\Sigma$, $P$)** <br>
События, образующие $\sigma$-алгебру могут быть как независимыми, так и зависимыми. Сама принадлежность событий к $\sigma$-алгебре не говорит о зависимости/независимости этих событий, нам также нужно учитывать набор событий. <br>

В тривиальной $\sigma$-алгебре всегда имеем $\Sigma = \{\varnothing, \Omega\} \implies$ события независимы, так как для них всегда выполняется $P(A \cap B) = P(A)\cdot P(B)$, а сами значения равны либо $P(A \cap B) = 0$, либо $P(A \cap B) = 1$<br>
Или в случае, когда изначальное пространтство замкнуто и детерминированнно, например, попадание в мишень, где изначально $\Omega = \{\omega\} \implies \Sigma = \{\varnothing, \Omega\} \implies$ события независимы <br>

Рассмотрим случаи, когда пространтство не ограничено. Пусть $A = \{2, 4, 6\}$ и $B = \{2\}$ события образующие $\sigma$-алгебру, тогда пространство исходов $\Omega = \{1, 2, 3, 4, 5, 6\}$. 
Имеем следующие непересекающиеся исходы событий:<br>
- $B = \{2\}$<br>
- $A \backslash B = \{4, 6\}$<br>
- $A^c = \{1, 3, 5\}$<br>

Так как в данном случаи мы рассматриваем полную $\sigma$-алгебру, тогда находим все пересечения событий:<br>
$\Sigma = \{\varnothing, \{2\}, \{4, 6\}, \{1, 3, 5\}, \{2, 4, 6\}, \{1, 2, 3, 5\}, \{1, 3, 4, 5, 6\}, \Omega\}$

Дальше проверяем зависимы ли события A и B, образующие $\sigma$-алгебру, для этого уточним, что рассматриваем честный кубик:<br>
- $P(B) = \frac{1}{6}$<br>
- $P(A) = \frac{1}{2}$<br>
- $P(A \cap B) = \frac{1}{6}$<br>

$$P(A) \cdot P(B) = \frac{1}{2} \cdot \frac{1}{6} = \frac{1}{12}$$
События являются независимыми, когда $P(A \cap B) = P(A)\cdot P(B)$, в данном случае $P(A \cap B) \neq P(A)\cdot P(B)$

**(c) События, имеющие одинаковую вероятность** <br>
Равенство вероятности событий не гарантирует зависимость или независимость данных событий. <br>
Независимость определяется $P(A \cap B) = P(A)\cdot P(B)$ <br>

Подбрасываем монетку, найдем зависимы ли события. Получаем<br>
$\Omega = \{O, P\}$
- $A = \{O\}$<br>
- $B = \{P\}$<br>

- $P(A) = \frac{1}{2}$<br>
- $P(B) = \frac{1}{2}$<br>

Обе вероятности равны, теперь проверяем независимы ли они:<br>
$$P(A \cap B) = 0$$
$$P(A)\cdot P(B) = \frac{1}{4} \neq 0 \implies \text{события зависимы}$$

Рассмотрим другой пример и проверим будут ли собятия с одинаковой вероятностью также зависимы как в примере выше. 

Две монетки подбрасываются, найдем зависимы ли события, если на обеих монетках выпал Орел. Получаем<br>
$\Omega = \{OO, OP, PO, PP\}$
- $A = \{OO, OP\}$<br>
- $B = \{OO, PO\}$<br>

- $P(A) = \frac{1}{2}$<br>
- $P(B) = \frac{1}{2}$<br>

Обе вероятности равны, теперь проверяем независимы ли они:<br>
$$P(A \cap B) = {OO} = \frac{1}{4}$$
$$P(A)\cdot P(B) = \frac{1}{4} = \frac{1}{4} \implies \text{события независимы}$$

Получается, что одинаковая вероятность не дает однозначного ответа зависимы ли события или нет.

• A — появление герба на первой монете;<br>

$A = \{ГГ, ГР\}$ <br>
$B= \{РГ, РР\} $<br>

#### 4. Два парохода должны подойти к одному и тому же причалу независимо друг от друга и равновозможно в течение суток. Определить вероятность того,что одному из них придется ожидать освобождения причала, если время стоянки первого парохода — 1 час, а второго — 2 часа.<br>

#### Ответ <br>
Пароходы идут к причалу в течении 24 часов, имеем отрезок от 0 до 24. Таки образом, мы имеем $$\Omega = [0; 24] \times [0;24]$$
Тогда пусть $x$ - время прихода первого парохода, $y$ - время прихода второго парохода, $(x,y) \in \Omega$<br>

Построим $\Omega$


```python
import matplotlib.pyplot as plt

# Plot
plt.figure(figsize=(8, 8))

plt.xlim(0, 24)
plt.ylim(0, 24)
plt.xlabel("Arrival time 1 ship")
plt.ylabel("Arrival time 2 ship")
plt.title("Ω")

plt.gca().set_aspect("equal", adjustable="box")
plt.grid(True, linestyle=':')

plt.show()
```

Так как параходы стоят на причале, то имеем два интервала при которых они могут встретиться и нам необходимо найти их пересечение: $[x, x+1]$ и $[y, y+2]$, тогда имеем два случая: <br>
1. Первый приходит не позже, чем уходит второй, то есть $y<x<y+2 \implies 0<x-y<2$<br>
2. Второй приходит не позже, чем уходит первый, то есть $x<y<x+1 \implies 0<y-x<1 \implies -1<x-y<0$<br>

Объединяем два случая и получаем, что $$-1<x-y<2$$
То есть точки $(x,y)$ лежат в области между двумя диагоналями $y=x+1$ и $y=x-2$. Добавим область пересечения на график:

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon

# Arrival times grid
x = np.linspace(0, 24, 400)
y = np.linspace(0, 24, 400)
X, Y = np.meshgrid(x, y)

# Overlap condition:
overlap = (X < Y + 2) & (Y < X + 1)

# Plot
plt.figure(figsize=(8, 8))

# Shade the overlap region
plt.contourf(X, Y, overlap, levels=[0, 0.5, 1], alpha=0.3)

# Boundary lines: y = x - 2  and  y = x + 1
plt.plot(x, x - 2, 'b--', linewidth=1)
plt.plot(x, x + 1, 'g--', linewidth=1)

plt.xlim(0, 24)
plt.ylim(0, 24)
plt.xlabel("Arrival time 1 ship")
plt.ylabel("Arrival time 2 ship")
plt.title("Ω")

plt.gca().set_aspect("equal", adjustable="box")
plt.grid(True, linestyle=':')

plt.show()
```

Таким образом, нас интересует зеленая область на графике, где параходы встретяться, то есть одному из них нужно будет ожидать пока отойдет другой. Нам необходимо найти ее площадь, для этого нужно вычесть из площади квадрата площади верхнего и нижнего треугольника.<br>
$$S_{\Omega} = 24 \times 24 = 576$$
Катеты верхнего треугольника равны 23 (так как верхняя граница $y=x+1$), тогда площадь верхнего треугольника равна:
$$S_{1} = \frac{23 \times 23}{2} = 264.5$$
Катеты нижнего треугольника равны 22 (так как верхняя граница $y=x-2$), тогда площадь нижнего треугольника равна:
$$S_{2} = \frac{22 \times 22}{2} = 242$$
Площадь искомой области равна:
$$S_{A} = S_{\Omega} - S_{1} - S_{2} = 576 - 264.5 - 242 = 69.5$$
Тогда вероятность попадания в эту область будет:
$$P(A) = \frac{S_{A}}{S_{\Omega}} = \frac{69.5}{576}$$
Ответ: $P(A) =\frac{69.5}{576}$

#### 5. Самолет, по которому ведется стрельба, состоит из трех различных по уязвимости частей:<br>
(a) Кабина летчика и двигатель<br>
(b) Топливные баки<br>
(c) Планер<br>
Для поражения самолета достаточно либо одного попадания в первую часть, либо двух попаданий во вторую, либо трех в третью. При попадании в самолет одного снаряда, снаряд с вероятностью p1 попадает в первую часть, с вероятностью p2 — во вторую, с вероятностью p3 — в третью. Попавшие снаряды распределяются по частям независимо друг от друга.<br>
Известно, что в самолет попало m снарядов. Найти условную вероятность P(A|m) события A – «Самолет поражен» – при m= 1,2,3,4.

#### Ответ <br>
Имеем 3 узвимые части:<br>
(a) Кабина летчика и двигатель -> 1 попадание<br>
(b) Топливные баки -> 2 попадания<br>
(c) Планер -> 3 попадания<br>

Вероятность попадания одним снаярдом равна $p_{1} + p_{2} + p_{3} = 1$<br>

Обозначим, что:<br>
$X_{1}$ - попадания в первую часть (a)<br>
$X_{2}$ - попадания во вторую часть (b)<br>
$X_{3}$ - попадания в третью часть (c)<br>

Самолет будет поражен, то есть событие А наступит при выполнении одного из условий: $X_{1} \geq 1$, $X_{2} \geq 2$, $X_{3} \geq 3$

Рассмотрим каждый m отдельно. <br>

**m=1**<br>
При $m=1$, событие А наступит только при попадании в часть а, то есть $$P(A|1) = p_{1}$$

**m=2**<br>
В данном случае имеем два случая поражения самолета: $X_{1} \geq 1$ либо $X_{2} \geq 2$, $X_{3}$ - не выполняется автоматически. 
Тогда либо выполняется условие a $$P(X_{1} \geq 1) = 1 - P(X_{1} = 0) = 1 - (1-p_{1})^{2} = 1 - (1-2p_{1}+p_{1}^{2}) = 2p_{1} - p_{1}^{2}$$
Либо выполняется условие b $$P(X_{2} \geq 2) = p_{2}^{2}$$

Объединяем условия: $$P(A|2) = 2p_{1} - p_{1}^{2} + p_{2}^{2}$$

**m=3**<br>
В этом случае имеем вероятность выполнения одного из всех трех условий: $X_{1} \geq 1$, $X_{2} \geq 2$, $X_{3} \geq 3$
Тогда проще найти случаи, при которых условие не выполняется и это: $X_{1} = 0$, $X_{2} = 1$, $X_{3} = 2$ <br>
В таком случае $$P(X_{1} = 0, X_{2} = 1, X_{3} = 2) = \frac{3!}{1!2!} \cdot p_{2} \cdot p_{3}^{2} = 3p_{2}p_{3}^{2}$$
Получаем: $$P(A|3) = 1 - P(X_{1} = 0, X_{2} = 1, X_{3} = 2) = 1 - 3p_{2}p_{3}^{2}$$

**m=4**<br>
В этом случае также имеем вероятность выполнения одного из всех трех условий: $X_{1} \geq 1$, $X_{2} \geq 2$, $X_{3} \geq 3$<br>
Пробуем найти случаи, при которых условие не выполняется:<br>
- Если $X_{1} = 0$, тогда $X_{2} + X_{3} = 4$ - при любой комбинации условие выполняется<br>
- Если $X_{2} = 0$, тогда $X_{1} + X_{3} = 4$ - при любой комбинации условие выполняется<br>
- Если $X_{3} = 0$, тогда $X_{1} + X_{2} = 4$ - при любой комбинации условие выполняется<br>

В таком случае получаем, что нет ни одной комбинации, при которой событие A не произошло бы: $$P(A|4) = 1$$

Ответ:<br>
$P(A|1) = p_{1}$,<br>
$P(A|2) = 2p_{1} - p_{1}^{2} + p_{2}^{2}$,<br>
$P(A|3) = 1 - 3p_{2}p_{3}^{2}$,<br>
$P(A|4) = 1$<br>

## 1.2 Случайный вектор и числовые характеристики

#### 1. Пусть

$$f_{\xi}(x,y) = \frac{e^{−2|y|}}{\pi(1 + x^2)}$$
Является ли данная функция плотностью распределения случайного вектора?

#### Ответ <br>
Чтобы функция $f_{\xi}(x,y)$ была плотность распределения случайного вектора $(x,y)$ необходимо, чтобы она отвечала двум условиям:<br>
1. Неотрицательность<br>
2. Нормирована (то есть интеграл по всей плоскости был равен 1)<br>
Рассмотим каждое из условий<br>

**Неотрицательность**<br>

Получается, что $e^{−2|y|} > 0, \forall y$, так как $e^{−2|y|}$ это экспонента, которая всегда положительна от любого действительного числа.<br>
Рассмотрим вторую часть функции: $\pi(1 + x^2)$, так как $x^2$, то любое значение $x$ - будет всегда положительно, остальные константы, поэтому верно выражение $\pi(1 + x^2) > 0, \forall x$

В таком случае обе части выражения всегда положительны, то есть верно: $$f_{\xi}(x,y)\geq 0, \forall(x,y)$$
Значит условие неотрицательности выполнено. 

**Нормирована**<br>
Функция: $$f_{\xi}(x,y) = \frac{e^{−2|y|}}{\pi(1 + x^2)}$$
Находим ее интеграл:
$$\iint f_{\xi}(x,y)\; \mathrm{d}x \mathrm{d}y = \int_{-\infty}^{\infty}\int_{-\infty}^{\infty} \frac{e^{−2|y|}}{\pi(1 + x^2)} \mathrm{d}x \mathrm{d}y$$
Так как имеем интеграл произведения двух переменных, можем найти каждый по отдельности. Первый интеграл: <br>
$$\int_{-\infty}^{\infty} \frac{\mathrm{d}x}{\pi(1 + x^2)}$$

Это стандартный интеграл:
$$\int_{-\infty}^{\infty} \frac{\mathrm{d}x}{(1 + x^2)} = \pi \implies \frac{1}{\pi} \times \pi = 1$$

Второй интеграл:
$$\int_{-\infty}^{\infty} e^{−2|y|}\mathrm{d}y = \int_{-\infty}^{0} e^{2y}\mathrm{d}y + \int_{0}^{\infty} e^{−2y}\mathrm{d}y $$
$$\int_{-\infty}^{0} e^{2y}\mathrm{d}y = \left[\frac{1}{2}e^{2y}\right]_{-\infty}^0 = \frac{1}{2} - 0 = \frac{1}{2}$$
$$\int_{0}^{\infty} e^{−2y}\mathrm{d}y  = \left[-\frac{1}{2}e^{-2y}\right]_{0}^{\infty} = 0 -(-\frac{1}{2}) = \frac{1}{2}$$

$$\int_{-\infty}^{\infty} e^{−2|y|}\mathrm{d}y = \int_{-\infty}^{0} e^{2y}\mathrm{d}y + \int_{0}^{\infty} e^{−2y}\mathrm{d}y = \frac{1}{2} + \frac{1}{2} = 1$$

Объединяем:
$$\iint f_{\xi}(x,y)\; \mathrm{d}x \mathrm{d}y = 1 \cdot 1 = 1$$

Значит функция нормирована. <br>

Так как оба условия выполняются, можно точно сказать, что $$f_{\xi}(x,y) = \frac{e^{−2|y|}}{\pi(1 + x^2)}$$
является плотностью распределения случайного вектора.

#### 5. Найти плотность распределения суммы двух независимых случайных величин ξ и η, если ξ∼Exp2 и η∼U0,1.

Получается плостность распределения СВ $\xi$ равна: 
$$f_{\xi}(u) = \begin{cases} 2e^{-2u}, & \quad u\geq 0 \\ 0, & \quad u<0 \end{cases}
$$ 

Плостность распределения СВ $\eta$ равна: 
$$f_{\eta}(z) = \begin{cases} 1, & \quad 0<z<1 \\ 0, & \quad else \end{cases}
$$ 

Переменные независимы и тогда плотность суммы $S = \xi + \eta$. По свертке:
$$f_{S}(x) = \int f_{\xi}(u)f_{\eta}(x-u) \mathrm{d}u$$

Плотность равномерной $\eta$ равна 1, но только на отрезке: 
$$ 0<\eta<1 \implies f_{\eta}(x-u) = 1 \implies 0< x-u <1 \implies x-1 < u <x$$

Получаем интервалы $u\in(x-1, x)$ и $f_{\xi}(u)$ не ноль только при $u\geq 0$, тогда наш рабочий интервал $u \in [max(0,x-1), x]$, тогда:
$$f_{S}(x) = \int_{max(0,x-1)}^{x} 2e^{-2u} \mathrm{d}u$$

При $x<0$ сумма не может быть отрицательной значит $f_{S}(x) = 0$ <br> 

При $0\leq x \leq 1$:<br> 
$$x \in [0, x]$$
$$f_{S}(x) = \int_{0}^{x} 2e^{-2u} \mathrm{d}u = -e^{-2x} - (-1) = 1 -e^{-2x}$$

При $x>1$:<br> 
$$x \in [x-1, x]$$
$$f_{S}(x) = \int_{x-1}^{x} 2e^{-2u} \mathrm{d}u = e^{-2(x-1)} - e^{-2x} = e^{-2x}(e^2 -1)$$

В итоге получаем
$$f_{S}(x) = \begin{cases} 0, & \quad x<0 \\ 1 -e^{-2x}, & \quad 0\leq x \leq 1 \\ e^{-2x}(e^2 -1), & \quad  x > 1 \end{cases}
$$
